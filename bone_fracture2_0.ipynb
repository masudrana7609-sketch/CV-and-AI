```python
import os
import random

test_images_path = "/content/drive/MyDrive/Colab Notebooks/bonefractureyolo/test/images"


random_image = random.choice(os.listdir(test_images_path))
random_image_path = os.path.join(test_images_path, random_image)

print("Random image selected:", random_image_path)
```

```python
from ultralytics import YOLO

model = YOLO("runs/detect/train/weights/best.pt")

results = model.predict(
    source=random_image_path,
    save=True,
    imgsz=640
)
```

```python
import matplotlib.pyplot as plt
import cv2

result_folder = results[0].save_dir
result_image_path = os.path.join(result_folder, random_image)


img = cv2.imread(result_image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 8))
plt.imshow(img)
plt.axis("off")
plt.title("YOLO Detection Result")
plt.show()
```
