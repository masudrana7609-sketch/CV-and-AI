```python
import os
import random

test_images_path = "/content/drive/MyDrive/Colab Notebooks/bonefractureyolo/test/images"

# pick a random image
random_image = random.choice(os.listdir(test_images_path))
random_image_path = os.path.join(test_images_path, random_image)

print("Random image selected:", random_image_path)
```

```python
from ultralytics import YOLO

model = YOLO("runs/detect/train/weights/best.pt")

results = model.predict(
    source=random_image_path,
    save=True,
    imgsz=640
)
```

```python
import matplotlib.pyplot as plt
import cv2

# Find the latest prediction folder automatically
result_folder = results[0].save_dir
result_image_path = os.path.join(result_folder, random_image)

# Load and show the predicted image
img = cv2.imread(result_image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6, 6))
plt.imshow(img)
plt.axis("off")
plt.title("YOLO Detection Result")
plt.show()
```

```python
from ultralytics import YOLO

# 1. Define the path to your best trained model weights
# This path is relative to the Google Colab environment after training.
BEST_MODEL_PATH = '/content/runs/detect/train/weights/best.pt'

# The path to your data.yaml file on Google Drive
DATA_YAML_PATH = '/content/drive/MyDrive/Colab Notebooks/bonefractureyolo/data.yaml'

# Load the best model
best_model = YOLO(BEST_MODEL_PATH)

# 2. Run validation on the test set
# We use 'split="test"' to tell YOLO to use the test folder defined in your data.yaml
metrics = best_model.val(data=DATA_YAML_PATH, split='test')

# 3. Print the key accuracy scores
print("\n--- YOLOv8 Bone Fracture Detection Accuracy ---")
print(f"Mean Average Precision (mAP) @ 50 IoU (mAP50): {metrics.box.map50:.4f}")
print(f"Mean Average Precision (mAP) @ 50-95 IoU (mAP50-95): {metrics.box.map:.4f}")
print("----------------------------------------------")
```
